```python
def turnGS(x, a=0.34, b=0.33, c=0.33):
    out = np.ndarray(shape=(x.shape[0:2]))
    out = np.zeros( x.shape[0:2] )
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            out[i,j]=a*x[i,j,0]+b*x[i,j,1]+c*x[i,j,2]
    return out
```

```python
gs_img = turnGS(img)
print(gs_img.shape)
gs_img
```
