# Tugas Mandiri 3

Masalah:
- Dua bola terhubung pegas
- Kedua bola dijatuhkan dari ketinggian berbeda, $h_1$ dan $h_2$
- Pegas hanya mampu meregang maksimum 10% dari kondisi awal

<br/>

Tujuan:
- Mendapatkan posisi kedua bola setiap waktu

<br/>

Asumsi:
- Percepatan gravitasi, $g = 9.8 m/s^2$
- Dua bola bermassa sama, $m_1 = m_2 = m$
- Konstanta pegas, $k$

<br/>

Teori:
- GLBB
    $$ \begin{equation} \tag{1} x_i = x_o + v.\Delta t \end{equation} $$
    
    $$ \begin{equation} \tag{2} v_i = v_o + a.\Delta t \end{equation} $$

- Tumbukan inelastik
    $$ \begin{equation} \tag{3} e = \left\lvert \frac{v_f}{v_i} \right\rvert \end{equation} $$

- Gaya gesek udara
    Gaya gesek udara secara sederhana dijabarkan dalam bentuk persamaan linear.
    $$ \begin{equation} \tag{4} F_{ges} = -b \vec{v} \end{equation} $$

    dengan $b$ adalah koefisien gaya gesek.

- Gaya pemulih pegas
    $$ \begin{equation} \tag{5} F_p = k \Delta l \end{equation}

<br/>
<br/>

Analisis:
- Saat bola dijatuhkan dari $H_o$, ketinggian bola berkurang, bola mengalami percepatan karena gravitasi. Ketinggian dan kecepatan benda saat di udara berlaku persamaan GLBB.
- Saat mencapai lantai, $H_i = 0$. Arah kecepatan perlu diubah tandanya. Nilai mutlak kecepatan berkurang karena faktor koefisien restitusi, $e = 0.8$.
- Gaya pemulih pegas dan gaya gravitasi berkontribusi pada percepatan setiap bola.
- Saat berada di udara, bola akan mengalami perlambatan karena faktor gesekan udara.
- Saat bola berada di puncak lokal ketika di udara, arah kecepatan akan berubah tanda karena faktor gravitasi. Dari persamaan GLBB, arah kecepatan akan otomatis berubah tanda.
- Bola disebut berhenti ketika $H_i = 0$ dan $v_i = 0$.

Analisis Persamaan Gerak:

**Kondisi 1**\
Saat kondisi $L'$ < $L$. Pegas membentuk sudut $\theta$ dari sumbu horizontal. Posisi bola A lebih tinggi dari B.

Tinjau gaya-gaya di A\
Arah x:
$$
    \begin{equation} \tag{1.1}
    \begin{split}
        \sum{} F_x = m_A a_x \\
        -F_{px} = m_A a_x \\
        -k \Delta l cos\theta = m_A a_x \\
        -\frac{k}{m_A} \Delta l cos\theta = a_x \\
    \end{split}
    \end{equation}
$$
Arah y:
$$
    \begin{equation} \tag{1.2}
    \begin{split}
        \sum{} F_y = m_A a_y \\
        F_{py} - F_{gA} = m_A a_y \\
        k \Delta l sin\theta - m_A g = m_A a_y \\
        \frac{k}{m_A} \Delta l sin\theta - g = a_y \\
    \end{split}
    \end{equation}
$$

<br>

Tinjau gaya-gaya di B\
Arah x:
$$
    \begin{equation} \tag{1.3}
    \begin{split}
        \sum{} F_x = m_B a_x \\
        F_{px} = m_B a_x \\
        k \Delta l cos\theta = m_B a_x \\
        \frac{k}{m_B} \Delta l cos\theta = a_x \\
    \end{split}
    \end{equation}
$$
Arah y:
$$
    \begin{equation} \tag{1.4}
    \begin{split}
        \sum{} F_y = m_B a_y \\
        -F_{py} - F_{gB} = m_B a_y \\
        -k \Delta l sin\theta - m_B g = m_B a_y \\
        -\left( \frac{k}{m_B} \Delta l sin\theta + g \right) = a_y \\
    \end{split}
    \end{equation}
$$
